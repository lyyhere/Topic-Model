```python
def get_ingredients(s):
    if s:
        result = []
        first_split = (s.replace(".\n", ".").split("."))
        for item in first_split:
            if (item.find(':') > 0):   # We're working with description
                result.append(item.strip("-").split(":")[0])
            elif (len(item) > 0):
                result.extend([ingredient.strip() for ingredient in item.split(",")])
        return ",".join(result)
    else:
        return []
```
